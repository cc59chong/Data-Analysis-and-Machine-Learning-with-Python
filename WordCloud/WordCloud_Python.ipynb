## 1. Basic one. Directly import text. 

```python
import wordcloud

# Create a wordcloud object and assign it to w.
w = wordcloud.WordCloud()

# Call the object and import text.
w.generate('BET mission is to drive efficient and reliable business practices through innovation, centralization, \
           and standardization in order to empower GSII to pursue their strategic priorities and achieve a center of excellence.\
           Business Excellence provides manage daily operational processes centered on business data management and business\
           operations management like data analytics, business operations, budget, vendor management, and training & development \
           to enable GSII to focus on core job responsibilities. We do partner with stakeholders on internal reporting projects\
           that speak to GSIIs value and Return on Investment. ')

# Save the generated wordcloud as 'output1.png' and save it to the current folder.
# w.to_file('output1.png')

# Show the wordcloud in Juypter notebook
import matplotlib.pyplot as plt

plt.axis("off")
plt.imshow(w)
plt.show()
```

## 5. Add the parameter 'scale' to improve clarity

## 6. Parameter 'stopword' -- Words that are not displayed.

```python
mk = imageio.imread("alice.png")
w = wordcloud.WordCloud(mask=mk)

w = wordcloud.WordCloud(background_color='white',
                        mask=mk,
                        scale=15,
                        stopwords={'Alice','said'}) # Don't dispaly 'Alice' and 'said'

f = open('alice_in_wonderland.txt')
txt = f.read()

w.generate(txt)

#w.to_file('output6.png')

plt.axis("off")
plt.imshow(w)
plt.show()
```

## 8. Use parameter 'interpolation' 

```python
mk = imageio.imread("alice.png")
w = wordcloud.WordCloud(mask=mk)

w = wordcloud.WordCloud(background_color='white',
                        mask=mk,
                        contour_width=5,
                        contour_color='blue')

f = open('alice_in_wonderland.txt')
txt = f.read()

w.generate(txt)

#w.to_file('output8.png')

plt.axis("off")
plt.imshow(w,interpolation='catrom') 
# The details of imterpolation: https://matplotlib.org/gallery/images_contours_and_fields/interpolation_methods.html?highlight=interpolation
plt.show()
```

## 9. Use custom colors

```python
mk = imageio.imread("alice.png")
w = wordcloud.WordCloud(mask=mk)

import random
# The function of custom color
def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)

w = wordcloud.WordCloud(background_color='white',
                        mask=mk,
                        max_words=2000,
                        contour_width=5,
                        contour_color='red')

f = open('alice_in_wonderland.txt')
txt = f.read()

w.generate(txt)

#w.to_file('output9.png')

plt.axis("off")
plt.imshow(w.recolor(color_func=grey_color_func, random_state=3)) 
plt.show()
```

## 10. Fill in the color according to the template

```python
from wordcloud import ImageColorGenerator

text = open('alice_in_wonderland.txt').read()

# Call the image of color template
mk = imageio.imread("alice_color.png")

w = wordcloud.WordCloud(background_color="white",
              mask=mk)

w.generate(text)

#use the 'ImageColorGenerator()' function to extract the colors of each part of the template image
image_colors = ImageColorGenerator(mk)

fig, axes = plt.subplots(1, 3)

# The left image shows the original wordcloud image
axes[0].imshow(w)

# The middle image shows the wordcloud image generated according to the template image color
# Use 'bilinear' interpolation method to display the color
axes[1].imshow(w.recolor(color_func=image_colors), interpolation="bilinear")

# The right image shows the template image
axes[2].imshow(mk)

for ax in axes:
    ax.set_axis_off()
plt.show()

# Recolor the wordcloud object as in the template image
# w_color = w.recolor(color_func=image_colors)

# w_color.to_file('output10.png')
```
